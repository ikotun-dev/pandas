```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#housing dataset 
housing_data = pd.read_csv('housing.csv')
housing_data.head()
```

```python
housing_data.tail()
```

```python
housing_data["ocean_proximity"] = (housing_data["ocean_proximity"] == "INLAND").astype(int)
#housing_data["ocen_proximity"] = (housing_data["ocean_proximity"] == "NEAR Bay").astype(int)
housing_data
```

```python
#groupby keyword 
housing_data.groupby('ocean_proximity').mean()['median_house_value']
```

```python
#check for missing values 
housing_data.isnull().sum()
```

```python
sns.lmplot(x='median_income', y='median_house_value', data=housing_data, fit_reg=True)
sns.lmplot(x='ocean_proximity', y='median_house_value', data=housing_data, fit_reg=True)
sns.lmplot(x='median_income', y='median_house_value', data=housing_data, fit_reg=True)
```
