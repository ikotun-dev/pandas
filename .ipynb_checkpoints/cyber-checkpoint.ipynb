```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
house_data = pd.read_csv('house_data.csv')
```

```python
house_data.head(15)
```

```python
house_data.tail(30)
```

```python
house_data.shape
```

```python
location_mapping= {}
for index, location in enumerate(house_data['location'].unique(), start=1):
    location_mapping[location] = index 
    
house_data['location'] = house_data['location'].map(location_mapping)
```

```python
city_mapping= {}
for index, city in enumerate(house_data['city'].unique(), start=1):
    city_mapping[city] = index 
    
house_data['city'] = house_data['city'].map(city_mapping)
```

```python
property_mapping= {}
for index, property in enumerate(house_data['property_type'].unique(), start=1):
    property_mapping[property] = index 
    
house_data['property_type'] = house_data['property_type'].map(property_mapping)
```

```python
house_data.corr()
```

```python
correlation_matrix = house_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0)
plt.show()
```
